```python
#importing packages from nltk library
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
#sample text document to apply our algorithm
data=open('President-Speech.txt').read()
```

```python
#defining function summary which consists of sequence of sentence scoring steps

def summary(dataset):
    result=[] #empty list to generate final result which will consist of sequence number, sentence and its coressponding score
    for number,sentence in enumerate(sent_tokenize(dataset)): # using sentence tokenization
        words_token=len(word_tokenize(sentence))  #isolating each word using word tokenization
        tagged=nltk.pos_tag(word_tokenize(sentence)) #using part of speech tagging as one of the factors in sentence scoring
        nn_noun=len([word for word,pos in tagged if pos in ['NN','NNP'] ]) #score nn_noun if the word belongs to noun category
        ners=nltk.ne_chunk(nltk.pos_tag(word_tokenize(sentence))) #using chunking
        nn_ners=len([chunk for chunk in ners if hasattr(chunk,'label')]) 
        score=(nn_noun + nn_ners)/float(words_token) #ultimate score formula for sentence scoring
        result.append((number,sentence,score)) #appending our empty list
    return result


    
        
        
        
        
        
        
        
    
```

Applying the function summary on our data

```python
summ=summary(data) 
```

```python
summ.sort(key=lambda x:x[2],reverse=True)
```
